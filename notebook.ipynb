```python
stat = [sport["TrainingCenterDatabase"]["Activities"]["Activity"]["Lap"] for sport in db["activities-Running"].find()]
```

```python
moy_time_coursses = []
for i in stat:
    try:
        moy_time_coursses.append(float(i["TotalTimeSeconds"])/3600)
    except:
        dis = 0.0
        for p in i:
            dis = dis + float(p["TotalTimeSeconds"])
        moy_time_coursses.append(dis/3600)
print(moy_time_coursses)
```

```python
moy_dis_coursses = []
for i in stat:
    try:
        dis = float(i["DistanceMeters"])/1000
        moy_dis_coursses.append(dis)
    except:
        dis = 0.0
        time = 0.0
        for p in i:
            dis = dis + float(p["DistanceMeters"])
        moy_dis_coursses.append((dis/1000))
print(moy_dis_coursses)
```

```python
moy_cal_coursses = []
for i in stat:
    try:
        moy_cal_coursses.append(i["Calories"])
    except:
        dis = 0.0
        for p in i:
            dis = dis + float(p["Calories"])
        moy_cal_coursses.append(dis)
print(moy_cal_coursses)
```

```python
moy_vit_coursses = []
for i in stat:
    try:
        dis = float(i["DistanceMeters"])/1000
        time = float(i["TotalTimeSeconds"])/3600
        moy_vit_coursses.append(dis/time)
    except:
        dis = 0.0
        time = 0.0
        for p in i:
            dis = dis + float(p["DistanceMeters"])
            time = time + float(p["TotalTimeSeconds"])
        moy_vit_coursses.append((dis/1000)/(time/3600))
print(moy_vit_coursses)
```

```python
moy_all_coursses = []
for i in stat:
    try:
        dis = float(i["DistanceMeters"])/1000
        time = float(i["TotalTimeSeconds"])/60
        moy_all_coursses.append(time/dis)
    except:
        dis = 0.0
        time = 0.0
        for p in i:
            dis = dis + float(p["DistanceMeters"])
            time = time + float(p["TotalTimeSeconds"])
        moy_all_coursses.append((time/60)/(dis/1000))
print(moy_all_coursses)
```

```python
long_lat_all_coursses = []
for i in stat:
    long_lat_coursses = []
    try:
        for i2 in i["Track"]["Trackpoint"]:
            long_lat_coursses.append([float(i2["LatitudeDegrees"]),float(i2["LongitudeDegrees"])])
    except:
        for p in i:
            for p2 in p["Track"]["Trackpoint"]:
                long_lat_coursses.append([float(p2["LatitudeDegrees"]),float(p2["LongitudeDegrees"])])
    
    long_lat_all_coursses.append(long_lat_coursses)
print(long_lat_all_coursses)
```

```python
alt_all_coursses = []
for i in stat:
    alt_coursses = []
    try:
        for i2 in i["Track"]["Trackpoint"]:
            alt_coursses.append(float(i2["AltitudeMeters"]))
    except:
        for p in i:
            for p2 in p["Track"]["Trackpoint"]:
                alt_coursses.append(float(p2["AltitudeMeters"]))
    alt_all_coursses.append(alt_coursses)
print(alt_all_coursses)
```

```python
BPM_all_coursses = []
for i in stat:
    BPM_coursses = []
    try:
        for i2 in i["Track"]["Trackpoint"]:
            BPM_coursses.append(float(i2["HeartRateBPM"]))
    except:
        try:
            for p in i:
                for p2 in p["Track"]["Trackpoint"]:
                    BPM_coursses.append(float(p2["HeartRateBPM"]))
        except:
            pass
    BPM_all_coursses.append(BPM_coursses)
print(BPM_all_coursses)
```

```python
moy_alt_all_coursses = []
for i in alt_all_coursses:
    if i:
        moy_alt_coursses = sum(i)/len(i)
    else:
        moy_alt_coursses = 0
    moy_alt_all_coursses.append(moy_alt_coursses)
print(moy_alt_all_coursses)
```
